```python
# Import libraries
import numpy as np
import pandas as pd
import os.path
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the patient data
_data = np.load('data/re_intubation.npy', allow_pickle=True).tolist()
patients = _data['patients']
print("Loaded!")
```

```python
# Select only true successful extubations and failures
success_bool = (patients['success']==True) & (patients['hospital_expire_flag']==0) # successful if never re-intubated and survived hospital stay
failure_bool = patients['failure']==True # failure if they were re-intubated within 48 hours
patients = patients[success_bool | failure_bool]

# Show some sample data
patients.head()
```

```python
# Print some statistics
print('Total number of patients: {}'.format(len(patients)))
print('Successful extubations: {} ({:.1f}%)'.format(sum(patients['success']), 100*sum(patients['success'])/len(patients)))
print('Extubation failures: {} ({:.1f}%)'.format(sum(patients['failure']), 100*sum(patients['failure'])/len(patients)))
```
